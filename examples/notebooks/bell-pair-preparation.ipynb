# Bell pair preparation

A Bell pair is the simplest entangled state — the foundational building block of every quantum communication protocol.

:::{important} Entanglement is the channel
Without the Bell pair there is no way to transmit information from Alice to Bob using just two classical bits. The entanglement is the *resource* that makes teleportation work.
:::

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2, 2)
qc.h(0)        # put qubit 0 in superposition
qc.cx(0, 1)    # entangle qubit 1 with qubit 0
qc.measure([0, 1], [0, 1])
```

:::{tip}
In the noiseless limit, measurement outcomes are perfectly correlated — both 00 or both 11, never mixed.
:::

```python
from qiskit_aer import AerSimulator
sim = AerSimulator()
job = sim.run(qc, shots=4096)
counts = job.result().get_counts()
```
